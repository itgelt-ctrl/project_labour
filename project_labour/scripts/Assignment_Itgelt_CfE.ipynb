Everything is type object, gotta make it right for analysis!

```python
# Rename columns English friendly
data.rename(columns={
    'Статистик үзүүлэлт': 'agegr',
    'Аймаг': 'province',
    'Хүйс': 'gndr'
}, inplace=True)

# Verify the column names
print( data.columns)
```

I will turn the gender from object to binary as male 0, female 1. And I will also convert province to str. And I will keep age group (agegr) variable as it is suitable for this analysis.


```python

data['gndr'] = data['gndr'].replace({'Эрэгтэй': 0, 'Эмэгтэй': 1})

# Verify the changes
print(data['gndr'].unique())
```

```python
# Convert 'province' to string
data['province'] = data['province'].astype('string')

# Verify the changes
print(data.dtypes)
```

### Let's use loop to turn columns 1992-2023 into a numeric variable.

```python
for year in range(1992, 2024):
    data[str(year)] = data[str(year)].replace({',': ''}, regex=True).astype(float)

# Verify the data types
print(data.dtypes)
```

Lets take a look at the data now. 

```python
display(data)
```

### Let's work on the missing values. 

```python
# Count missing values in each column
print(data.isnull().sum())
```

There are quite a lot of missing values. This is a problem we can's address and some regions seems to have skipped years to collect data. We have to find a way to analyse our data. For this assignment, I will only concentrate on capital city and country data conduct required analysis.  

```python
# I will change the names of values into English as: 
data['province'] = data['province'].replace({
    'Улсын дүн': 'country',
    'Улаанбаатар': 'capital'
})

# Let's take a loot
print(data['province'].unique())


```

Now we will be concentrating on country and capital values. 

```python
# Let's filter the province of interest, in this case its country and capital. 
filtered_data = data[data['province'].isin(['country', 'capital'])]

# Display the filtered df. 
display(filtered_data)


```

### It seems great. Now we have all the capital and country level labour population data ready for analysis. 

I would like my analysis to concentrate on self determined most active age group of 20-64. I will be pruning age groups before and after this subset.


```python
# Filtering the observations of the right age group. 
filtered_data = filtered_data[filtered_data['agegr'].isin(['Бүгд','20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64'])]

filtered_data['agegr'] = filtered_data['agegr'].replace({
    'Бүгд': 'all',
})
display(filtered_data)
```

```python
# Let's save this amazing data. 
file_path = 'data/filtered_data.csv'

filtered_data.to_csv(file_path, index=False)
```

### Visualisations over time for Timeseries. 

```python
import matplotlib.pyplot as plt
os.makedirs('outputs', exist_ok=True)
# filtering for a 20-24 and country level
subset = filtered_data[(filtered_data['agegr'] == '20-24') & (filtered_data['province'] == 'country')]

years = subset.columns[3:]  # assuming years start from the fourth column
values = subset.iloc[0, 3:]  # first row of the filtered data for years

plt.figure(figsize=(12, 6))
plt.plot(years, values, marker='o', color='b', label='20-24, country')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Population Trend (20-24, country) ')
plt.xticks(rotation=45)
plt.legend()


#Save it
plt.savefig('outputs/population_trend_20-24_country.png')


```

```python
# Let's plot multiple age groups over time for country.
plt.figure(figsize=(12, 6))
age_groups = ['20-24', '25-29', '30-34']  

for age in age_groups:
    subset = filtered_data[(filtered_data['agegr'] == age) & (filtered_data['province'] == 'country')]
    values = subset.iloc[0, 3:]
    plt.plot(years, values, marker='o', label=age)

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Population Trends by Age Group (Country)')
plt.xticks(rotation=45)
plt.legend()

#Save it
plt.savefig('outputs/Population_Trends_by_Age_Group_(Country).png')
```
